```python
tasks = []   # list to store tasks

while True:
    print("\n--- TO DO APP ---")
    print("1. Add Task")
    print("2. View Tasks")
    print("3. Mark Task as Done")
    print("4. Delete Task")
    print("5. Exit")

    choice = input("Enter your choice: ")

    # 1. Add Task
    if choice == "1":
        task_name = input("Enter task name: ")
        task = {
            "title": task_name,
            "done": False
        }
        tasks.append(task)
        print("Task added successfully")

    # 2. View Tasks
    elif choice == "2":
        if len(tasks) == 0:
            print("No tasks available")
        else:
            for i in range(len(tasks)):
                status = "Done" if tasks[i]["done"] else "Pending"
                print(i + 1, ".", tasks[i]["title"], "-", status)

    # 3. Mark Task as Done
    elif choice == "3":
        for i in range(len(tasks)):
            status = "Done" if tasks[i]["done"] else "Pending"
            print(i + 1, ".", tasks[i]["title"], "-", status)

        num = int(input("Enter task number to mark done: "))
        tasks[num - 1]["done"] = True
        print("Task marked as done")

    # 4. Delete Task
    elif choice == "4":
        for i in range(len(tasks)):
            print(i + 1, ".", tasks[i]["title"])

        num = int(input("Enter task number to delete: "))
        tasks.pop(num - 1)
        print("Task deleted")

    # 5. Exit
    elif choice == "5":
        print("Exiting To Do App")
        break

    else:
        print("Invalid choice")
```
