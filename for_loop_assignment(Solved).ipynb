## Question 9 (Easy)

**Title:** Count Even and Odd Numbers

### Problem Statement:
Take N as input and count how many even and odd numbers are there between 1 and N.

### Approach:
Use two counters and check with modulo operator.


```python
# write your code here
N = int(input("Enter N: "))
even_n = 0
odd_n = 0

for i in range(1, N + 1):
    if i % 2 == 0:
        even_n += 1
    else:
        odd_n += 1

print("Even:", even_n)
print("Odd:", odd_n)
```

## Question 17 (Hard)

**Title:** Prime Numbers Between 1 and N

### Problem Statement:
Take N as input and print all prime numbers between 1 and N.

### Approach:
Use nested loops and a flag variable for prime checking.


```python
# write your code here
N = int(input())

for num in range(2, N + 1):
    is_prime = True

    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break

    if is_prime:
        print(num)
```

## Question 19 (Hard)

**Title:** Number Pattern

### Problem Statement:
Take N as input and print a right-angled number pattern using nested loops.

Example:
1
12
123

### Approach:
Use an outer loop for rows and an inner loop for printing numbers.


```python
# write your code here
N = int(input())

for i in range(1, N + 1):
    for j in range(1, i + 1):
        print(j, end="")
    print()
```
