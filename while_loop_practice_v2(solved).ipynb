Q1. Print numbers from 1 to 20 using while loop.

Approach:
- Initialize a variable
- Use while condition
- Increment inside loop

```python
#write answer here
n = 1
while n<=20:
  print(n)
  n=n+1
```

Q4. Find sum of numbers from 1 to n.

Approach:
- Take input n
- Add numbers in loop

```python
#write answer here
n = int(input())
total = 0
i = 1

while i <= n:
    total += i
    i += 1

print(total)
```

Q5. Reverse a number using while loop.

Approach:
- Extract last digit
- Build reversed number

```python
#write answer here

n = int(input())
rev = 0

while n > 0:
    digit = n % 10
    rev = rev * 10 + digit
    n = n // 10

print(rev)
```

Q6. Count digits in a number.

Approach:
- Divide number by 10 repeatedly
- Count iterations

```python
#write answer here
n = int(input())
count = 0

while n > 0:
    n = n // 10
    count += 1

print(count)
```

Q7. Check if a number is palindrome.

Approach:
- Reverse number
- Compare with original

```python
#write answer here
n = int(input())
original = n
rev = 0

while n > 0:
    digit = n % 10
    rev = rev * 10 + digit
    n = n // 10

if rev == original:
    print("Palindrome")
else:
    print("Not Palindrome")
```

Q8. Guess the number game.

Approach:
- Use while loop
- Continue until correct guess

```python
#write answer here
n = int(input())
original  = 6
while n != original:
  n = int(input())
print("Correct")
```
